```python
from pyspark.sql import SparkSession
spark=(SparkSession.builder.getOrCreate())
```

```python
emp = [
("001",101,"John Doe",30,"Male",50000,"2015-01-01"),
("002",101,"Jane Smith",25,"Female",45000,"2016-02-15"),
("003",102,"Bob Brown",35,"Male",55000,"2014-05-01"),
("004",102,"Alice Lee",28,"Female",48000,"2017-09-30"),
("005",103,"Jack Chan",40,"Male",60000,"2013-04-01"),
("006",103,"Jill Wong",32,"Female",52000,"2018-07-01"),
("007",101,"James Johnson",42,"Male",70000,"2012-03-15"),
("008",102,"Kate Kim",29,"Female",51000,"2019-10-01")
]

columns = [
"employee_id",
"department_id",
"name",
"age",
"gender",
"salary",
"hire_date"
]
```

```python
emp_1=spark.createDataFrame(data=emp,schema=columns)
```

```python
emp_1.show()
```

```python
emp_1.select("employee_id","name","salary").show()
```

```python
emp_1.where("salary>60000").show()
```

```python
from pyspark.sql.functions import col,cast,expr
emp_3=emp_1.withColumn("salary",col("salary").cast("double"))
```

```python
emp_4=emp_3.withColumn("tax",col("salary")*0.1)
```

```python
from pyspark.sql.functions import when
emp_5=emp_4.withColumn("gender_short",when(col("gender")=="Male","M").when(col("gender")=="Female","F").otherwise(None))
```

```python
from pyspark.sql.functions import regexp_replace 
emp_6=emp_5.withColumn("new_name",regexp_replace(col("name"),"Lee","Li"))
```

```python
from pyspark.sql.functions import to_date

emp_7=emp_6.withColumn("hire_date",to_date(col("hire_date"),"yyyy-MM-dd"))
```
